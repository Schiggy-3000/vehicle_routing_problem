```python
import requests
import json
import urllib
import urllib.request
```

```python
def create_data():
  """Creates the data."""
  data = {}

  # API key:
  # Google Cloud Console > APIs and services > Enable "Distance Matrix API"
  # Google Cloud Console > APIs and services > Credentials > Create credentials > API key
  data['API_key'] = 'ADD_YOUR_API_KEY_HERE'
  data['addresses'] = ['3610+Hacks+Cross+Rd+Memphis+TN', # depot
                       '1921+Elvis+Presley+Blvd+Memphis+TN',
                       '149+Union+Avenue+Memphis+TN',
                       '1034+Audubon+Drive+Memphis+TN'
                      ]
  return data
```

```python
def build_address_str(addresses):
    # Build a pipe-separated string of addresses
    # Apparently, the Google API requires this format
    address_str = ''
    for i in range(len(addresses) - 1):
        address_str += addresses[i] + '|'
    address_str += addresses[-1]
    return address_str
```

```python
#####################
###### Testing ######
#####################

data = create_data()
addresses = build_address_str(data["addresses"])
addresses
```

```python
def send_request(origin_addresses, dest_addresses, API_key):
  """ Calculate the distance and duration between each pair of addresses."""
  
  # Google Distance Matrix API documentation:
  # https://developers.google.com/maps/documentation/distance-matrix/distance-matrix
  
  output_format = 'json' # json or xml
  #departure_time = ... # Either departure_time or arrival_time may be set, but not both.
  #arrival_time = ... # Either departure_time or arrival_time may be set, but not both.
  #avoid = ... # tolls, highways, ferries, indoor
  language = 'en' # Language in which to return results.
  mode = 'driving' # driving, walking, bicycling, transit.
  region = 'us' # us, ch, ... This represents the region code, necessary if, for instance, an address is ambiguous.
  #traffic_model = 'best_guess' # best_guess, pessimistic, optimistic. For calculating travel times. Only when departure_time is set.
  #transit_mode = ... # bus, subway, train, tram. Only used if mode=transit.
  #transit_routing_preference = ... # less_walking, fewer_transfers. Only used if mode=transit.
  units = 'metric' # metric or imperial.
  origins = build_address_str(origin_addresses)
  destinations = build_address_str(dest_addresses)

  api_key = API_key
  base_url = 'https://maps.googleapis.com/maps/api/distancematrix/'
  request_url = base_url + output_format + \
                '?language=' + language + \
                '&mode=' + mode + \
                '&region=' + region + \
                '&units=' + units + \
                '&origins=' + origins + \
                '&destinations=' + destinations + \
                '&key=' + api_key

  jsonResult = urllib.request.urlopen(request_url).read()
  response = json.loads(jsonResult)
  
  return response

```

```python
#####################
###### Testing ######
#####################

data = create_data()
origin_addresses = data["addresses"]
dest_addresses = data["addresses"]
API_key = data["API_key"]

# This will calculate the distance and duration between each pair of addresses
response = send_request(origin_addresses, dest_addresses, API_key)
response
```

```python
def build_distance_matrix(response):

  # Calculate the distance (in meters) between each pair of addresses and store in a matrix
  distance_matrix = []
  for row in response['rows']:
    row_list = [row['elements'][j]['distance']['value'] for j in range(len(row['elements']))]
    distance_matrix.append(row_list)
  return distance_matrix
```

```python
#####################
###### Testing ######
#####################

data = create_data()
origin_addresses = data["addresses"]
dest_addresses = data["addresses"]
API_key = data["API_key"]

# Calculate the distance (in meters) between each pair of addresses
response = send_request(origin_addresses, dest_addresses, API_key)
distance_matrix = build_distance_matrix(response)
distance_matrix
```

```python
def main():
  """Entry point of the program"""
  
  data = create_data()
  origin_addresses = data["addresses"]
  dest_addresses = data["addresses"]
  API_key = data["API_key"]

  # This will calculate the distance and duration between each pair of addresses
  response = send_request(origin_addresses, dest_addresses, API_key)
  distance_matrix = build_distance_matrix(response)
  print(distance_matrix)
  
```

```python
if __name__ == '__main__':
  main()
```
